```python
# Langkah 1: Impor modul yang diperlukan
import openpyxl
import os
import pandas as pd

# Langkah 2: Tentukan folder input dan output
input_folder = "C:/Users/Alif/OneDrive - uinjkt.ac.id/Kuliah UIN/PKL/BMKG/Data Iklim xlxs/2024"  # Ganti dengan path folder Anda
output_folder = "C:/Users/Alif/OneDrive - uinjkt.ac.id/Kuliah UIN/PKL/BMKG/Data Iklim csv/2024"  # Ganti dengan path folder tujuan

# Buat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Langkah 3: Konversi file .xlsx menjadi .csv
for filename in os.listdir(input_folder):
    if filename.endswith('.xlsx'):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename.replace('.xlsx', '.csv'))
        
        # Membaca file Excel dan menyimpannya sebagai CSV
        try:
            print(f"Converting {filename}...")
            df = pd.read_excel(input_path)
            df.to_csv(output_path, index=False)
            print(f"File saved as: {output_path}")
        except Exception as e:
            print(f"Error processing {filename}: {e}")

print("Konversi selesai.")
```

```python
# Langkah 1: Impor modul yang diperlukan
import os
import pandas as pd

# Langkah 2: Tentukan folder input dan output
input_folder = 'C:/Users/Alif/OneDrive - uinjkt.ac.id/Kuliah UIN/PKL/BMKG/Data Iklim csv/2024'  # Ganti dengan path folder Anda
output_folder = 'C:/Users/Alif/OneDrive - uinjkt.ac.id/Kuliah UIN/PKL/BMKG/Data Iklim csv - Data Only/2024'  # Ganti dengan path folder tujuan

# Buat folder output jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Langkah 3: Filter baris pada file CSV
for filename in os.listdir(input_folder):
    if filename.endswith('.csv'):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        # Membaca file CSV dan menyaring baris
        try:
            print(f"Processing {filename}...")
            df = pd.read_csv(input_path)
            
            # Hapus baris 1 sampai 7 dan baris setelah 40
            filtered_df = df.iloc[6:40]
            
            # Simpan file CSV yang telah difilter
            filtered_df.to_csv(output_path, index=False, header=False)  # Simpan tanpa header jika perlu
            print(f"File saved as: {output_path}")
        except Exception as e:
            print(f"Error processing {filename}: {e}")

print("Proses selesai.")
```
