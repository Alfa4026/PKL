### **Analisis Hubungan Curah Hujan dan Kecepatan Angin terhadap Variasi Temperatur Rata-rata di Wilayah Tangerang, Kawasan Bandara Soekarno-Hatta**

---

#### **Latar Belakang**  
Kawasan Bandara Soekarno-Hatta di Tangerang memiliki karakteristik cuaca yang unik, di mana variabilitas curah hujan, kecepatan angin, dan suhu dapat berdampak signifikan pada operasi penerbangan, termasuk keberangkatan, pendaratan, serta keselamatan penerbangan. Dengan demikian, penting untuk memahami bagaimana curah hujan dan kecepatan angin memengaruhi variasi temperatur rata-rata di kawasan ini.

## Analisis residual

Setelah mendapatkan model terbaik dengan melakukan pemilihan model menggunakan stepwise, selanjutnya akan dilakukan analisis residual untuk melihat apakah model regresi yang dipilih memenuhi asumsi-asumsi dari regresi linear atau tidak.

### Asumsi normalitas

Melihat histogram dan qqplot

````r
install.packages("olsrr")
````

````r
library(olsrr)
resid1 <- residuals(model3)
resid2 <- rstandard(model3)
hist(resid2, xlab = "residual", main = "")
ols_plot_resid_qq(model3)
````

Uji hipotesis normalitas:

-   H0: residual mengikuti distribusi normal

-   H1: residual tidak mengikuti distribusi normal


````r
ols_test_normality(model2)
````

### Asumsi Homoskedastisitas

````r
ols_plot_resid_fit(model3)
plot(fitted(model3), rstandard(model3))
````

**uji Breusch-Pagan**

Uji hipotesis variansi konstan:

-   H0: variansi residual konstan (homoskedastisitas)

-   H1: variansi residual tidak konstan (heteroskedastisitas)

````r
# Install and load the lmtest package
# install.packages("lmtest")
library(lmtest)

# Perform the Breusch-Pagan test
bptest(model3)
````

### Asumsi Independen

**Dengan uji Durbin-Watson**

Uji hipotesis variansi konstan:

-   H0: residual independent

-   H1: residual tidak independent

````r
durbinWatsonTest(model3)
````

**Dengan Uji Breusch-Godfrey**

````r
# Uji Breusch-Godfrey
bg_test <- bgtest(model3, order = 3)  # Order = jumlah lag yang diuji
print(bg_test)
````

### **Penjelasan Konsep Lag dalam Permasalahan Uji Breusch-Godfrey**

Dalam analisis model regresi yang dilakukan, uji Breusch-Godfrey digunakan untuk memeriksa **autokorelasi residual**, yaitu hubungan antara residual saat ini (\(e_t\)) dengan residual pada periode sebelumnya (\(e_{t-1}, e_{t-2}, \dots\)). Konsep **lag** dalam konteks ini mengacu pada "periode keterkaitan" residual dalam data.

Misalnya:
- **Lag 1 (\(e_{t-1}\))**: Residual saat ini (\(e_t\)) dibandingkan dengan residual satu periode sebelumnya.  
- **Lag 2 (\(e_{t-2}\))**: Residual saat ini (\(e_t\)) dibandingkan dengan residual dua periode sebelumnya.  
- **Lag 3 (\(e_{t-3}\))**: Residual saat ini (\(e_t\)) dibandingkan dengan residual tiga periode sebelumnya.  

Dalam uji yang dilakukan, Breusch-Godfrey menguji keberadaan autokorelasi residual **hingga lag ke-3**.

---

### **Hasil Uji Breusch-Godfrey**

Berikut adalah hasil uji Breusch-Godfrey:
```
Breusch-Godfrey test for serial correlation of order up to 3

data:  model3
LM test = 6.7078, df = 3, p-value = 0.08182
```

**Penjelasan:**
1. **LM Test**: Statistik uji sebesar **6.7078** menunjukkan tingkat autokorelasi yang dihitung berdasarkan hubungan residual hingga lag ke-3.  
2. **df (Derajat Kebebasan)**: Derajat kebebasan sebesar **3** karena uji dilakukan hingga lag ke-3.  
3. **p-value**: Nilai **p-value = 0.08182** menunjukkan probabilitas hasil uji ini terjadi jika hipotesis nol (\(H_0\)) benar.

---

### **Interpretasi**

1. **Hipotesis Uji Breusch-Godfrey**:
   - **Hipotesis Nol (\(H_0\))**: Tidak ada autokorelasi residual hingga lag ke-3.  
   - **Hipotesis Alternatif (\(H_a\))**: Ada autokorelasi residual hingga lag ke-3.

2. **Hasil Uji**:
   - Karena **p-value = 0.08182 > 0.05**, maka **gagal menolak hipotesis nol (\(H_0\))**.  
   - Ini berarti **tidak ada bukti statistik yang cukup untuk menyimpulkan adanya autokorelasi residual** hingga lag ke-3.

3. **Kesimpulan Praktis**:
   - Residual dari model **independen** hingga lag ke-3, sehingga asumsi independensi residual dalam model ini terpenuhi.

---

### **Relevansi Lag dalam Permasalahan**

Dalam analisis model:
- Uji Breusch-Godfrey menguji autokorelasi residual untuk memastikan bahwa residual pada waktu \(t\) **tidak dipengaruhi** oleh residual pada waktu sebelumnya (\(t-1, t-2, \dots\)).
- Hasil menunjukkan bahwa **tidak ada autokorelasi signifikan** hingga lag ke-3, sehingga model valid dari sisi asumsi independensi residual.

# Model Dengan Interaksi

1. Tn (Temperatur Minimum) dan Tx (Temperatur Maksimum)

    Interaksi: Suhu minimum dan maksimum bisa saling mempengaruhi. Perbedaan besar antara suhu maksimum dan minimum dapat memberikan informasi tentang fluktuasi suhu harian yang bisa memengaruhi suhu rata-rata (Tavg).

2. Tn (Temperatur Minimum) dan RH_avg (Kelembapan Rata-rata)

    Interaksi: Kelembapan yang tinggi dapat mempengaruhi pengurangan penurunan suhu pada malam hari. Interaksi ini mungkin berhubungan dengan pengaruh kelembapan terhadap suhu malam hari dan bagaimana hal ini mempengaruhi suhu rata-rata.

3. Tx (Temperatur Maksimum) dan ss (Lamanya Penyinaran Matahari)

    Interaksi: Lamanya penyinaran matahari berhubungan langsung dengan suhu maksimum yang tercatat. Interaksi antara keduanya sangat relevan, karena durasi penyinaran matahari dapat memengaruhi suhu maksimum dan pada gilirannya berpotensi mempengaruhi Tavg.

4. ff_avg (Kecepatan Angin Rata-rata) dan ff_x (Kecepatan Angin Maksimum)

    Interaksi: Angin yang lebih kuat dapat mengubah distribusi suhu dan mempengaruhi suhu rata-rata dengan cara mengurangi efek pemanasan di atmosfer atau mempercepat pendinginan.

5. RH_avg (Kelembapan Rata-rata) dan RR (Curah Hujan)

    Interaksi: Curah hujan dan kelembapan seringkali berkorelasi tinggi. Curah hujan tinggi sering kali terjadi ketika kelembapan relatif tinggi, yang bisa menurunkan suhu sekitar (efek pendinginan) dan mempengaruhi suhu rata-rata.

6. ff_avg (Kecepatan Angin Rata-rata) dan ddd_car (Arah Angin Terbanyak)

    Interaksi: Arah angin dapat mempengaruhi suhu di wilayah tersebut. Interaksi antara kecepatan angin rata-rata dan arah angin bisa mempengaruhi bagaimana angin mendistribusikan suhu di area tertentu.
