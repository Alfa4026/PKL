# Outlier

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ganti 'your_column_name' dengan nama kolom yang ingin Anda periksa
column_name = 'TN'

plt.figure(figsize=(10, 6))
sns.boxplot(x=data[column_name])
plt.title(f'Boxplot for Outlier Detection in {column_name}')
plt.xlabel(column_name)
plt.show()
```

```python
# Menentukan fungsi untuk mendeteksi outlier menggunakan IQR
def detect_outliers(data):
    # Menyaring kolom numerik saja
    numeric_cols = data.select_dtypes(include=['number']).columns
    
    # Menyimpan indeks outlier
    outlier_indices = []
    
    for col in numeric_cols:
        # Hitung Q1 (25%) dan Q3 (75%)
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Tentukan batas bawah dan atas untuk outlier
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        
        # Temukan indeks yang merupakan outlier
        outlier_indices.extend(data[(data[col] < lower_limit) | (data[col] > upper_limit)].index)
    
    # Menghapus duplikat indeks outlier
    outlier_indices = list(set(outlier_indices))
    
    # Menghapus data yang merupakan outlier
    data_no_outlier = data.drop(index=outlier_indices)
    
    return data_no_outlier

# Menghapus outlier
data_no_outlier = detect_outliers(data)

# Menampilkan jumlah data setelah menghapus outlier
print(f"Jumlah data setelah menghapus outlier: {data_no_outlier.shape[0]}")
```

```python
import pandas as pd

# Fungsi untuk imputasi missing value pada data numerik dengan interpolasi dan median
def impute_missing_values(data_no_outlier):
    # Menangani variabel numerik dengan interpolasi terlebih dahulu
    numeric_cols = data_no_outlier.select_dtypes(include=['number']).columns
    for col in numeric_cols:
        # Lakukan interpolasi linier untuk mengisi missing value
        data_no_outlier[col] = data_no_outlier[col].interpolate(method='linear')
        
        # Jika ada missing value yang tidak terisi karena interpolasi (misalnya jika nilai missing terlalu panjang), isi dengan median
        if data_no_outlier[col].isnull().sum() > 0:
            median_value = data_no_outlier[col].median()  # Mengambil median
            data_no_outlier[col].fillna(median_value, inplace=True)  # Imputasi dengan median
    
    return data_no_outlier

# Imputasi missing values
data_imputed = impute_missing_values(data_no_outlier)

# Menampilkan data setelah imputasi
print(data_imputed.isnull().sum())  # Mengecek apakah masih ada missing values
```
