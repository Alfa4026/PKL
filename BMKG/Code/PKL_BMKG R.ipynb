### **Analisis Hubungan Curah Hujan dan Kecepatan Angin terhadap Variasi Temperatur Rata-rata di Wilayah Tangerang, Kawasan Bandara Soekarno-Hatta**

---

#### **Latar Belakang**  
Kawasan Bandara Soekarno-Hatta di Tangerang memiliki karakteristik cuaca yang unik, di mana variabilitas curah hujan, kecepatan angin, dan suhu dapat berdampak signifikan pada operasi penerbangan, termasuk keberangkatan, pendaratan, serta keselamatan penerbangan. Variasi temperatur rata-rata dapat memengaruhi kondisi atmosfer lokal, yang memengaruhi densitas udara dan performa pesawat. Dengan demikian, penting untuk memahami bagaimana curah hujan dan kecepatan angin memengaruhi variasi temperatur rata-rata di kawasan ini.  

Penelitian ini diharapkan dapat memberikan kontribusi pada prakiraan cuaca mikro di sekitar bandara, mitigasi dampak perubahan iklim, serta membantu meminimalisir potensi kecelakaan penerbangan akibat kondisi cuaca ekstrem.

---

#### **Tujuan Penelitian**  
1. Menganalisis pengaruh curah hujan terhadap variasi temperatur rata-rata di kawasan Bandara Soekarno-Hatta.  
2. Menilai kontribusi kecepatan angin maksimum dan rata-rata terhadap perubahan temperatur rata-rata.  
3. Memberikan rekomendasi terkait mitigasi risiko operasional di bandara berdasarkan analisis hubungan variabel cuaca.

---

#### **Variabel yang Digunakan**
1. **Curah hujan (RR):** Menunjukkan intensitas curah hujan yang memengaruhi pendinginan atmosfer.  
2. **Kecepatan angin maksimum (ff_x):** Menggambarkan kecepatan angin tertinggi yang berpotensi memengaruhi sirkulasi udara lokal.  
3. **Kecepatan angin rata-rata (ff_avg):** Menunjukkan pola pergerakan angin yang memengaruhi distribusi suhu di kawasan bandara.  
4. **Temperatur rata-rata (Tavg):** Merupakan parameter utama untuk mengukur kondisi termal harian di kawasan tersebut.

---

#### **Metodologi Penelitian**
1. **Pengumpulan Data**:  
   - Data meteorologi (RR, ff_x, ff_avg, Tavg) dari stasiun BMKG di Bandara Soekarno-Hatta selama periode tertentu (misalnya 5 tahun terakhir).  

2. **Pembersihan Data**:  
   - Menghilangkan nilai outlier dan data kosong untuk memastikan kualitas analisis.  

3. **Analisis Statistik**:  
   - **Korelasi Pearson/Spearman** untuk mengevaluasi hubungan antara curah hujan, kecepatan angin, dan temperatur rata-rata.  
   - **Regresi Linear Berganda** untuk memodelkan pengaruh variabel independen (RR, ff_x, ff_avg) terhadap variabel dependen (Tavg).  

4. **Visualisasi Data**:  
   - Membuat grafik scatter plot untuk menggambarkan hubungan antarvariabel.  
   - Heatmap untuk menunjukkan kekuatan korelasi antarvariabel.  
   - Tren temporal untuk melihat perubahan parameter cuaca secara berkala.  

---

#### **Hasil yang Diharapkan**  
1. **Hubungan Kuantitatif Antarvariabel:**  
   - Mengidentifikasi seberapa besar curah hujan dan kecepatan angin berkontribusi terhadap perubahan temperatur rata-rata.  

2. **Model Prediktif:**  
   - Menghasilkan model regresi untuk memprediksi temperatur rata-rata berdasarkan curah hujan dan kecepatan angin.  

3. **Rekomendasi Operasional:**  
   - Strategi mitigasi risiko cuaca untuk operasional bandara, seperti pengaturan jadwal penerbangan pada kondisi cuaca tertentu.  

---

## Analisis residual

Setelah mendapatkan model terbaik dengan melakukan pemilihan model menggunakan stepwise, selanjutnya akan dilakukan analisis residual untuk melihat apakah model regresi yang dipilih memenuhi asumsi-asumsi dari regresi linear atau tidak.

### Asumsi normalitas

Melihat histogram dan qqplot

```r
install.packages("olsrr")
```

```r
library(olsrr)
resid1 <- residuals(model2)
resid2 <- rstandard(model2)
hist(resid2, xlab = "residual", main = "")
ols_plot_resid_qq(model2)
```

Uji hipotesis normalitas:

-   H0: residual mengikuti distribusi normal

-   H1: residual tidak mengikuti distribusi normal


```r
ols_test_normality(model2)
```

- Uji Shapiro-Wilk, Kolmogorov-Smirnov, dan Anderson-Darling tidak   memberikan bukti yang cukup untuk menolak bahwa residual mengikuti distribusi normal.
- Namun, uji Cramer-von Mises memberikan hasil yang bertentangan dengan ketiga uji lainnya, dengan p-value yang sangat kecil, yang menunjukkan bahwa residual mungkin tidak mengikuti distribusi normal.

### Asumsi Homoskedastisitas

```r
ols_plot_resid_fit(model2)
plot(fitted(model2), rstandard(model2))
```

**uji Breusch-Pagan**

Uji hipotesis variansi konstan:

-   H0: variansi residual konstan (homoskedastisitas)

-   H1: variansi residual tidak konstan (heteroskedastisitas)

```r
# Install and load the lmtest package
# install.packages("lmtest")
library(lmtest)

# Perform the Breusch-Pagan test
bptest(model2)
```

Karena p-value sangat kecil (0.001151), Anda harus menolak hipotesis nol dan mempertimbangkan adanya heteroskedastisitas dalam model. Ini berarti variansi residual tidak konstan.

### Asumsi Independen

**Dengan uji Durbin-Watson**

Uji hipotesis variansi konstan:

-   H0: residual independent

-   H1: residual tidak independent

```r
durbinWatsonTest(model2)
```

p-value = 0 yang lebih kecil dari 0.05 menunjukkan bahwa kita menolak hipotesis nol (H0) dan menerima hipotesis alternatif (H1), yang berarti ada bukti kuat bahwa residual tidak independen dan terdapat autokorelasi dalam model regresi.

Karena residual model tidak homoskedastisitas dan residual tidak independen maka selanjutnya akan dilakukan transformasi terhadap variabel dependen atau TAVG.

```r
library(lmtest)
# Uji Breusch-Godfrey
bg_test <- bgtest(model2, order = 680)  # Order = jumlah lag yang diuji
print(bg_test)
```
