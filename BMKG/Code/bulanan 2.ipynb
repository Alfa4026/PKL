#### Ubah Tipe Data

```python
# Exclude columns that should not be converted to float
exclude_columns = ['DDD_CAR', 'TANGGAL']
data = data.apply(lambda col: col.replace('-', None).astype(float) if col.name not in exclude_columns and col.dtypes == 'object' else col)
data['DDD_CAR'] = data['DDD_CAR'].astype('category')
data['TANGGAL'] = pd.to_datetime(data['TANGGAL'], format='%d-%m-%Y', dayfirst=True)
```

```python

value_to_replace = [8888, 9999, 8888.0, 9999.0]

for col in data.columns:
    if data[col].dtype.kind in {'i', 'f'}:  # Kolom numerik (integer atau float)
        data.loc[:, col] = data[col].replace(value_to_replace, None).astype(float)
    elif data[col].dtype.kind == 'O':  # Kolom object (string/kategorikal)
        data.loc[:, col] = data[col].replace(value_to_replace, None)
    elif data[col].dtype.kind == 'M':  # Kolom datetime
        # Tidak mengganti nilai untuk kolom datetime
        pass
```

```python
data['DDD_CAR'] = data['DDD_CAR'].replace('S ', 'S')
data['DDD_CAR'] = data['DDD_CAR'].replace('W ', 'W')
data['DDD_CAR'] = data['DDD_CAR'].replace('E ', 'E')
data['DDD_CAR'] = data['DDD_CAR'].replace('N ', 'N')
```

#### Missing Value

```python
data.isnull().sum() 
```

```python
missing_percentage = data.isnull().mean() * 100
print(missing_percentage)
```

```python
data.shape
```

```python
# Imputasi untuk variabel dengan missing value rendah menggunakan median
data['TN'] = data['TN'].fillna(data['TN'].median())
data['TX'] = data['TX'].fillna(data['TX'].median())
data['TAVG'] = data['TAVG'].fillna(data['TAVG'].median())
data['RH_AVG'] = data['RH_AVG'].fillna(data['RH_AVG'].median())
data['SS'] = data['SS'].fillna(data['SS'].median())
data['DDD_X'] = data['DDD_X'].fillna(data['DDD_X'].mode()[0])  # Mengisi dengan modus

# Imputasi dengan rata-rata bulanan untuk variabel dengan missing value signifikan (seperti RR)
data['RR'] = data['RR'].fillna(data.groupby(data['TANGGAL'].dt.month)['RR'].transform('mean'))

# Imputasi dengan interpolasi untuk variabel lainnya jika diinginkan
data['FF_AVG'] = data['FF_AVG'].interpolate()

# Tampilkan data setelah imputasi
print("Data Setelah Imputasi:")
data.head(40)
```

```python
# Tambahkan kolom bulan
data['BULAN'] = data['TANGGAL'].dt.to_period('M')

# Agregasi per bulan
data_bulanan = data.groupby('BULAN').agg({
    'TN': 'min',  # Temperatur minimum per bulan
    'TX': 'max',  # Temperatur maksimum per bulan
    'TAVG': 'mean',  # Rata-rata temperatur per bulan
    'RH_AVG': 'mean',  # Rata-rata kelembapan per bulan
    'RR': 'mean',  # Rata-rata curah hujan harian per bulan
    'SS': 'mean',  # Rata-rata penyinaran matahari harian per bulan
    'FF_X': 'max',  # Kecepatan angin maksimum per bulan
    'DDD_X': lambda x: x.mode()[0] if not x.mode().empty else None,  # Modus arah angin
    'FF_AVG': 'mean',  # Rata-rata kecepatan angin per bulan
    'DDD_CAR': lambda x: x.mode()[0] if not x.mode().empty else None  # Modus arah angin terbanyak
}).reset_index()

# Tampilkan hasil agregasi bulanan
print("\nData Setelah Agregasi Bulanan:")
data_bulanan
```

## **Agregasi Data Bulanan**

Proses ini bertujuan untuk mengubah data yang memiliki interval harian menjadi data dengan interval bulanan, agar dapat dianalisis dalam cakupan waktu yang lebih luas. Berikut adalah langkah-langkah yang dilakukan dalam agregasi data ini:

---

### **1. Menambahkan Kolom Bulan**
- Setelah kolom tanggal dikonversi menjadi datetime, kita menambahkan kolom baru yang berisi informasi bulan menggunakan `dt.to_period('M')`.  
- Kolom `bulan` ini digunakan sebagai acuan untuk agregasi berdasarkan bulan.

---

### **2. Agregasi Per Bulan**
Data dikelompokkan berdasarkan kolom `bulan` menggunakan fungsi `groupby('bulan')`. Setelah pengelompokan, masing-masing kolom dihitung dengan fungsi agregasi yang sesuai:

| **Variabel**  | **Fungsi Agregasi**                                       | **Penjelasan**                                                       |
|---------------|-----------------------------------------------------------|----------------------------------------------------------------------|
| **TN**        | `min()`                                                   | Mengambil temperatur minimum per bulan.                             |
| **TX**        | `max()`                                                   | Mengambil temperatur maksimum per bulan.                            |
| **TAVG**      | `mean()`                                                  | Menghitung rata-rata temperatur per bulan.                          |
| **RH_AVG**    | `mean()`                                                  | Menghitung rata-rata kelembapan per bulan.                          |
| **RR**        | `mean()`                                                  | Menghitung rata-rata curah hujan harian dalam sebulan.              |
| **SS**        | `mean()`                                                  | Menghitung rata-rata lamanya penyinaran matahari harian per bulan.  |
| **FF_X**      | `max()`                                                   | Mengambil kecepatan angin maksimum per bulan.                       |
| **DDD_X**     | `lambda x: x.mode()[0] if not x.mode().empty else None`   | Mengambil modus arah angin saat kecepatan maksimum.                 |
| **FF_AVG**    | `mean()`                                                  | Menghitung rata-rata kecepatan angin per bulan.                     |
| **DDD_CAR**   | `lambda x: x.mode()[0] if not x.mode().empty else None`   | Mengambil modus arah angin terbanyak dalam sebulan.                 |

---

### **3. Hasil Agregasi**
- Hasil akhirnya adalah DataFrame baru yang berisi data yang sudah teragregasi per bulan.  
- Setiap baris mewakili satu bulan tertentu, dengan kolom-kolom yang berisi nilai hasil agregasi (minimum, maksimum, rata-rata, atau modus) sesuai kategori masing-masing variabel.  

---

### **4. Alasan Penggunaan Mean untuk RR dan SS**
- **Curah Hujan (RR)**: Menggunakan rata-rata memberikan gambaran rata-rata curah hujan harian selama sebulan, yang relevan untuk analisis cuaca harian.  
- **Lamanya Penyinaran Matahari (SS)**: Rata-rata waktu penyinaran harian memberikan gambaran pola harian tanpa pengaruh jumlah hari dalam bulan tertentu.
